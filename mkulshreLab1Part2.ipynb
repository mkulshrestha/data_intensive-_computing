```r
ndata=read.csv("FluViewPhase2Data-3/WHO_NREVSS_Clinical_Labs.csv", TRUE, ",", skip=1)
#the cvs data is read and stored in ndata


pasted_value =(sprintf("%d%02d", ndata$YEAR,ndata$WEEK))

#sprintf will store the formatted data value in pasted_value

hdata=data.frame( ndata$TOTAL.B,ndata$TOTAL.A)

#data frames are created in the hdata based on parameters


hdata=t(hdata)

#transpose of hdata taken in hdata for plotting a stacked barpolt

barplot(as.matrix(hdata),main="Influenza positive tests reported by CDC US Clinincal Lab \nNational Summary, season 2018-2019",ylab="No. of +ve specimens",xlab="WEEKS",names.arg=pasted_value,col=c("green","yellow"), las=2,cex.names=0.5)
par(new=T)

#To plot multiple dataframes. Graphical parameters are set.


plot(ndata$PERCENT.POSITIVE , axes=FALSE ,type="l",ylab="",xlab="")
axis(4,las=1,at=seq(0,25,by=5))
lines(ndata$PERCENT.A, type="l",col="yellow",lty="dashed")
lines(ndata$PERCENT.B, type="l",col="green",lty="dashed")

mtext("Percent positive", side=4, line=1)

legend("topleft",legend=c("percent positive","percent A", "percent B","A","B"),
       col=c("black","yellow", "green","yellow","green"),cex=0.8,lty=c(1,3,3,0,0),pch=c(NA,NA,NA,15,15))
# dev.off()
```

```r
data7=read.csv("FluViewPhase2Data-3/WHO_NREVSS_Public_Health_Labs.csv",header=TRUE,",",skip=1)
# data7
# calculating the sum of 6 columns to find the cumulative positive specimens reported
A=sum(data7$A..2009.H1N1.)
A_H3=sum(data7$A..H3.)
A_subtype=sum(data7$A..Subtyping.not.Performed.)
B=sum(data7$B)
BVic=sum(data7$BVic)
Byam=sum(data7$BYam)
H3N2v=sum(data7$H3N2v)

labels=c("A 2009 H1N1", "A H3N2", "A subtyping not performed", "B unknown lineage", "B Victoria", "B yamagata")
#A single vector array is stored in labels


inp=data.frame(A,A_H3,A_subtype,B,BVic,Byam)
inp=t(inp)  #transposing to make a column
#taking transpose of inp to plot stacked bar plot.

pie(inp,radius=0.69, label=labels,col=rainbow(6), cex=0.5)

legend("topleft",legend=c("A 2009 H1N1", "A H3N2", "A subtyping not performed", "B unknown lineage", "B Victoria", "B yamagata"),
       col=rainbow(6),
       cex=0.8,
       lty=c(0,0,0,0,0,0),pch=c(15,15,15,15,15,15))
# dev.off()
```

```r
library(usmap)
library(maptools)
library(ggplot2)

data_map=read.csv("StateDatabySeason58_57.csv", header = TRUE)

level=(data_map$ACTIVITY.LEVEL)



data_map$ACTIVITY.LEVEL=as.numeric(gsub("Level ", "", data_map$ACTIVITY.LEVEL))



data_map$STATENAME=fips(data_map$STATENAME)
# data_map


#---fucntion to read inout from user in order to plot
#---heat map for the specific year
readweek <- function()
{ 
  n <- readline(prompt="Enter an integer bwtween 1 and 52 : ")
 
  return(n)
}

week=readweek()

readyear <- function()
{ 
  n <- readline(prompt="Enter season 2017-18 or 2018-19 : ")
  
  return(n)
}

year=readyear()



dat=data_map[which( data_map$STATENAME!="NA" & data_map$WEEK==week & data_map$SEASON==year),]

# dat
d=data.frame(fips=dat$STATENAME, level=dat$ACTIVITY.LEVEL)
# data frames are stored in d.

plot_usmap(data = d , values = "level", lines = "black",labels=TRUE, label_color="black" ) + 
labs(title = "A Weekly Influenza Surveillance Report Prepared by the Influenza Division \n
Influenza-Like Illness (ILI) Activity Level Indicator Determined by Data Reported to ILINet")+
  scale_fill_continuous(name = "LEVEL OF ACTIVITY", label = scales::comma,low="green", high="red") + 
  theme(legend.position = "right")
# dev.off()



# newdata
```
