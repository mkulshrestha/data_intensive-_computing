# PROBLEM 1 -  Define two synthetic vectors of data representing sales over 12 months for 2 items. 

```r
sales1<-c(12,14,16,29,30,45,19,20,16, 19, 34, 20)
sales2<-rpois(12,34) #POISSON DISTRIBUTION
# jpeg("part1.jpg")
par(bg="cornsilk")
plot(sales1, col="blue", type="o", ylim=c(0,100), xlab="Month", ylab="Sales" )
title(main="Sales by Month")
lines(sales2, type="o", pch=22, lty=2, col="red")
grid(nx=NA, ny=NULL)
legend("topright", inset=.05, c("Sales1","Sales2"), fill=c("blue","red"), horiz=TRUE)
# dev.off()
```

#### NORMAL DISTRIBUTION-12 numbers with mean=25 and std deviation=10

```r
sales1<-c(12,14,16,29,30,45,19,20,16, 19, 34, 20)
sales2<-rnorm(12,25,10) 
# jpeg("part2.jpg")


par(bg="cornsilk")
plot(sales1, col="blue", type="o", ylim=c(0,100), xlab="Month", ylab="Sales" )
title(main="Sales by Month")
lines(sales2, type="o", pch=22, lty=2, col="red")
grid(nx=NA, ny=NULL)
legend("topright", inset=.05, c("Sales1","Sales2"), fill=c("blue","red"), horiz=TRUE)
# dev.off()
```

```r
aapl1<-read.csv("AAPL.csv")
fb1<-read.csv("FB.csv")
# jpeg("part6.jpg")
par(bg="cornsilk")
plot(aapl1$Adj.Close, col="blue", type="o", ylim=c(125,250), xlab="Days", ylab="Price" )
lines(fb1$Adj.Close, type="o", pch=23, lty=2, col="red")
legend("topright", inset=.05, c("Apple","Facebook"), fill=c("red","blue"), horiz=TRUE)

hist(aapl1$Adj.Close, col=rainbow(8))
# dev.off()
```

The plot above compares the share prices of FACEBOOK and APPLE respectively over the duration of last year. clearly 
FACEBOOK attained a higher all time high of 230 around 160th trade day. whereas APPLE's all time high during the last year was capped around 220 around 120th trade day.

The Histogram shows the price Vs #of days the price was inside a range. For example the price was about 140-150 for a total of 5 days, similarly price was 150-160 for liitle more than 20 days e and so on. It is evident from the histogram that the price was in the range 180-190 for the maximum number of days i.e. around 58.
